```python
from probabilistic_classifier.experiment import multiclass_probabilistic_classifier_experiment
import numpy as np

import logging

logging.basicConfig(filename='multiclass_feature_size.log', filemode='w', format='%(levelname)s - %(asctime)s - %(message)s', datefmt='%d-%b-%y %H:%M:%S', level=logging.INFO)
```

```python
prime = 101
data_range = 2
num_of_samples = 800000

hidden_size_arr = [256, 256, 256]
lr = 0.0001

num_of_outer_iteration = 20
num_of_inner_iteration = 50
batch_size = 4096

para_param, priv_param = 1, 1
beta_arr, alpha_arr = [0, 1], [2, 3, 4]

feature_size = 7
```

```python
weight = np.ones((feature_size, 1))
dataset_feature_size = (2 * (para_param + priv_param - 1) + 3) * feature_size + 1
x_idx, y_idx, z_idx = list(range(feature_size + 1)), list(range(2 * feature_size + 1, dataset_feature_size)), list(range(feature_size + 1, 2 * feature_size + 1))
curr_ldr = multiclass_probabilistic_classifier_experiment(prime, data_range, num_of_samples, weight, feature_size, beta_arr, alpha_arr, para_param, priv_param, x_idx, y_idx, z_idx, hidden_size_arr, lr, num_of_outer_iteration, num_of_inner_iteration, batch_size)
logging.info('feature size: {}, ldr: {}'.format(feature_size, curr_ldr))
```
